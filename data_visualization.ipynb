```python

# ================================================================
# COVID-19 Data Visualization
# Prepared by: Darrah Borinaga
# Date: March_20_2025
# Purpose: Analyze and visualize COVID-19 case counts by province
# ================================================================

import matplotlib.pyplot as plt
import numpy as np

# Three_Provinces_with_Highest_Covid19
provinces = ["Canada", "Ontario", "Quebec"]
cases = [4669364, 1617673, 1341086]
percentages = [61.21, 21.21, 17.58]

# Square_format
fig, ax = plt.subplots(figsize=(8, 8), facecolor='white')

# Bars_colors
colors = ['red', 'blue', 'green']

# Bar_chart
bars = ax.bar(provinces, cases, color=colors, edgecolor='black', linewidth=1.2)

# Labels_above_bars
for bar, case, percentage in zip(bars, cases, percentages):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
            f'{case:,} ({percentage}%)', ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Title
ax.set_title("COVID-19 Case Counts by Province", fontsize=14, fontweight='bold', pad=20, color='darkblue')
ax.set_xlabel("Province", fontsize=12, fontweight='bold', labelpad=10, color='black')
ax.set_ylabel("Total Cases", fontsize=12, fontweight='bold', labelpad=10, color='black')

# Format x-axis
ax.set_xticks(range(len(provinces)))
ax.set_xticklabels(provinces, fontsize=10, rotation=10, fontweight='bold', color='black')

# Format y-axis
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))
ax.tick_params(axis='y', labelsize=10, colors='black')

# background
ax.set_facecolor('lightgray')
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Footer text
plt.figtext(0.1, 0.01, "Prepared by: Darrah_Borinaga_TMU_2025", ha="center",
            fontsize=8, fontweight='bold', color="black", backgroundcolor="black", bbox=dict(facecolor='none', edgecolor='white', boxstyle='round,pad=0.1'))

# Display the chart
plt.show()
```
